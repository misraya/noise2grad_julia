```julia
include("layers.jl")
include("helper.jl")
include("data_loader.jl")
include("loss.jl")
include("model.jl")
include("noise.jl")
```

```julia
train_ground_truth_path = "../Noise_datasets/train/ground_truth/" #x
train_reference_clean_path = "../Noise_datasets/train/reference_clean_image/" #y
test_ground_truth_path = "../Noise_datasets/test/ground_truth"
```

```julia
#display some images
test_set = collect(load(string(test_ground_truth_path ,"/",f)) for f in read_img_names_in_dir(test_ground_truth_path)[1:10]) 
test_set
```

```julia
test_set = TestDataset(test_ground_truth_path, 2)
```

```julia
train_set = TrainDataset(train_ground_truth_path, train_reference_clean_path, 2, limit=true) 
#true limits the instances to first 100, for fast debugging
```

```julia
model = N2G(8)
```

```julia
#forward pass, no training

for epoch in 1:2
    epoch_iters = 0
    epoch_loss_grad, epoch_loss_denoise, epoch_loss_total = 0, 0, 0
    

    for (x,y) in train_set.data
        epoch_iters += 2
        
        loss_grad, loss_denoise, loss_total = model(x, y)
        epoch_loss_grad+=loss_grad
        epoch_loss_denoise+=loss_denoise
        epoch_loss_total+=loss_total
        
    end
    
    println("\nEnd of epoch ", epoch)
    println("Average loss_grad of epoch ", epoch_loss_grad/epoch_iters)
    println("Average loss_denoise of epoch ", epoch_loss_denoise/epoch_iters)
    println("Average loss_total of epoch ", epoch_loss_total/epoch_iters)
    
    #model.update()

end

#similar result is expected between epochs since optimization/updating weights is not implemented yet
```

```julia
#evaluate on test set
total_psnr = 0
total_ssim = 0

for x_batch in test_set.data
    
    #try out different noises
    x_batch = additive_gaussian_noise(x_batch, false)
    #x_batch = poisson_noise(x_batch)
    #x_batch = speckle_noise(x_batch)
    denoised_batch = model.net(x_batch)[4]
    total_psnr += sum(psnr(x_batch,denoised_batch))
    total_ssim += sum(ssim(x_batch,denoised_batch))
end
    
avg_psnr = total_psnr / test_set.length
avg_ssim = total_ssim / test_set.length
println("Avg. PSNR on Test Set: ", avg_psnr) #the greater the better, goes to Inf.
println("Avg. SSIM on Test Set: ", avg_ssim) #between [0,1]. closer to 1 is better
```
